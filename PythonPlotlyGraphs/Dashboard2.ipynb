```python
import plotly as py
import plotly.plotly as pl
import plotly.graph_objs as go 
import numpy as np 
import pandas as pd
from plotly import tools

py.offline.init_notebook_mode(connected=True)
suicide_df=pd.read_csv('master.csv')

suicide_year_1985 = suicide_df[(suicide_df['year'] >= 1985) & (suicide_df['year'] < 1995)]
suicide_year_1995 = suicide_df[(suicide_df['year'] >= 1995) & (suicide_df['year'] < 2005)]
suicide_year_2005 = suicide_df[(suicide_df['year'] >= 2005) & (suicide_df['year'] < 2015)]
suicide_year_2015 = suicide_df[suicide_df['year'] >= 2015]

suicide_df_male = suicide_df[suicide_df['sex'] == 'male']
suicide_df_female = suicide_df[suicide_df['sex'] == 'female']

#male by year
suicide_year_1985_male =  suicide_year_1985[suicide_year_1985['sex'] == 'male']
suicide_year_1995_male =  suicide_year_1995[suicide_year_1995['sex'] == 'male']
suicide_year_2005_male =  suicide_year_2005[suicide_year_2005['sex'] == 'male']

#female by year
suicide_year_1985_female =  suicide_year_1985[suicide_year_1985['sex'] == 'female']
suicide_year_1995_female =  suicide_year_1995[suicide_year_1995['sex'] == 'female']
suicide_year_2005_female =  suicide_year_2005[suicide_year_2005['sex'] == 'female']

#age groups
suicide_df_15 = suicide_df[suicide_df['age'] == '15-24 years']
suicide_df_25 = suicide_df[suicide_df['age'] == '25-34 years']
suicide_df_35 = suicide_df[suicide_df['age'] == '35-54 years']
suicide_df_55 = suicide_df[suicide_df['age'] == '55-74 years']
suicide_df_75 = suicide_df[suicide_df['age'] == '75+ years']

#age/sex 
suicide_df_15_female = suicide_df_15[suicide_df['sex'] == 'female']
suicide_df_25_female = suicide_df_25[suicide_df['sex'] == 'female']
suicide_df_35_female = suicide_df_35[suicide_df['sex'] == 'female']
suicide_df_55_female = suicide_df_55[suicide_df['sex'] == 'female']
suicide_df_75_female = suicide_df_75[suicide_df['sex'] == 'female']

suicide_df_15_male = suicide_df_15[suicide_df['sex'] == 'male']
suicide_df_25_male = suicide_df_25[suicide_df['sex'] == 'male']
suicide_df_35_male = suicide_df_35[suicide_df['sex'] == 'male']
suicide_df_55_male = suicide_df_55[suicide_df['sex'] == 'male']
suicide_df_75_male = suicide_df_75[suicide_df['sex'] == 'male']

#group 1
x = ['male', 'female']
y = [suicide_df_male['suicides_no'].sum(), suicide_df_female['suicides_no'].sum()]

traceSex = go.Bar(
    x = x,
    y = y, 
)

x = ['1985 - 1994', '1995 - 2004', '2005 - 2014', ]
y = [suicide_year_1985_male['suicides_no'].sum(), suicide_year_1995_male['suicides_no'].sum(), suicide_year_2005_male['suicides_no'].sum()]

#group 2
traceSexMale = go.Bar(
x = x,
    y=y,
     marker=dict(
        color='rgb(10,10,225)',
        line=dict(
            color='rgb(5,5,250)',
            width=1.5,
        )
    ),
)

y = [suicide_year_1985_female['suicides_no'].sum(), suicide_year_1995_female['suicides_no'].sum(), suicide_year_2005_female['suicides_no'].sum()]

traceSexFemale = go.Bar(
x = x,
    y=y,
     marker=dict(
        color='rgb(225,10,225)',
        line=dict(
            color='rgb(250,5,250)',
            width=1.5,
        )
    ),
)

#group 3
suicidesGroup = suicide_df.groupby('year')
x = ['1987', '1988', '1989', '1990', '1991', '1992', '1993', '1994', '1995', '1996', '1997', '1998', '1999', '2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014']
traceSuicidesLine = go.Scatter(
x = x, 
    y = suicidesGroup['suicides_no'].count()
)

#group 4
xAge = ['15 - 24', '25 - 34', '35 - 54', '55 - 74', '75+']
yAge = [suicide_df_15['suicides_no'].sum(), suicide_df_25['suicides_no'].sum(), suicide_df_35['suicides_no'].sum(), suicide_df_55['suicides_no'].sum(), suicide_df_75['suicides_no'].sum()]

traceByYear = go.Bar(
x = xAge,
    y=yAge,
)

traceAgeMale = go.Bar(
x=xAge,
    y=[suicide_df_15_male['suicides_no'].sum(), suicide_df_25_male['suicides_no'].sum(), suicide_df_35_male['suicides_no'].sum(), suicide_df_55_male['suicides_no'].sum(), suicide_df_75_male['suicides_no'].sum()],
    marker=dict(
        color='rgb(10,10,225)',
        line=dict(
            color='rgb(5,5,250)',
            width=1.5,
        )
    ),
)

traceAgeFemale = go.Bar(
x=xAge,
    y=[suicide_df_15_female['suicides_no'].sum(), suicide_df_25_female['suicides_no'].sum(), suicide_df_35_female['suicides_no'].sum(), suicide_df_55_female['suicides_no'].sum(), suicide_df_75_male['suicides_no'].sum()],
     marker=dict(
        color='rgb(225,10,225)',
        line=dict(
            color='rgb(250,5,250)',
            width=1.5,
        )
    ),
)



layout = go.Layout(
    autosize=False,
    width=1000,
    height=1000,
    margin=go.layout.Margin(
        l=100,
        r=50,
        b=50,
        t=50,
        pad=10
    ),
    paper_bgcolor='#7f7f7f',
    plot_bgcolor='#c7c7c7'
)



fig = tools.make_subplots(rows = 4, cols=2, specs=[[{}, {}],[{}, {}], [{},{}], [{'colspan' : 2}, None]], subplot_titles=('Total Suicides By Sex',  'Total Suicides by age group', 'Male Suicides over years', 'Female Suicides over years', 'Suicides by age group (Male)', 'Suicides by age group (Female)', 'Suicides Over Time'))
fig.append_trace(traceSex, 1, 1)
fig.append_trace(traceByYear, 1, 2)
fig.append_trace(traceSexMale, 2, 1)
fig.append_trace(traceSexFemale, 2, 2)
fig.append_trace(traceAgeMale, 3, 1)
fig.append_trace(traceAgeFemale, 3, 2)
fig.append_trace(traceSuicidesLine, 4, 1)

fig['layout'].update(title='Suicide Rates Overview 1985 to 2016 - from Kaggle Datasets')

py.offline.plot(fig, filename = 'DanielBrandtDashboard2.html')
```
